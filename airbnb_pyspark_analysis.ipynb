```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import IntegerType, FloatType, DateType
```

```python
spark = SparkSession.builder.appName("property_analysis").getOrCreate()
```

```python
raw_df = spark.read.parquet("gs://airbnb_listing_analysis/raw_parquet")
```

```python
raw_df.show()
```

```python
agg_df=raw_df.groupBy("city").agg(count("*").alias("total_listings"),avg("review_scores_rating").\
                           alias("average_rating")).orderBy(col("total_listings").desc(),col("average_rating").desc())
```

```python
agg_df.show()
```

```python
raw_df.createOrReplaceTempView('airbnb_listings')
```

```python
query1 = spark.sql("""select listing_id,host_id,district,city from airbnb_listings""")
```
